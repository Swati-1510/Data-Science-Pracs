# **Simple Linear Regression -Global Superstore**

# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
```

# Load the dataset

```python
# Load the dataset
try:
    # Use 'latin1' encoding for compatibility with this dataset
    df = pd.read_csv('Datasets/Global_Superstore2.csv', encoding='latin1')
    print("Dataset 'Global_Superstore.csv' loaded successfully.")
except FileNotFoundError:
    print("Error: 'Datasets/Global_Superstore.csv' not found. Make sure it is in the same directory.")
    df = pd.DataFrame() # Create empty dataframe to avoid errors

# Display the first few rows to confirm it loaded correctly
if not df.empty:
    df.head()
```

# Data analysis

```python
# Describe data for relevant columns
if not df.empty:
    print(df[['Sales', 'Profit']].describe())
```

# Data Distribution Plot

```python
# Data distribution for our target variable 'Profit'
if not df.empty:
    plt.title('Profit Distribution Plot')
    sns.distplot(df['Profit'])
    plt.show()
```

# Check the relationship between Profit and Sales

```python
# Relationship between Profit and Sales
if not df.empty:
    plt.scatter(df['Sales'], df['Profit'], color = 'lightcoral')
    plt.title('Profit vs Sales')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.box(False)
    plt.show()
```

# Split the dataset into dependent/independent variables
Sales (X) is the independent variable

Profit (y) is dependent on Sales

```python
if not df.empty:
    X = df[['Sales']]  # Independent (needs to be a DataFrame)
    y = df[['Profit']] # Dependent (needs to be a DataFrame)
```

# Split data into Train/Test sets

```python
if not df.empty:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Train the regression model

```python
# Regressor model
if not df.empty:
    regressor = LinearRegression()
    regressor.fit(X_train, y_train)
```

# Predict the result

```python
# Prediction result
if not df.empty:
    y_pred_test = regressor.predict(X_test)     # predicted value of y_test
    y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

# Plot the training and test results

## Plot training set data vs predictions

```python
# Prediction on training set
if not df.empty:
    plt.scatter(X_train, y_train, color = 'lightcoral')
    plt.plot(X_train, y_pred_train, color = 'firebrick')
    plt.title('Profit vs Sales (Training Set)')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.legend(['Pred(y_train)', 'y_train'], title = 'Prediction/Actual', loc='best', facecolor='white')
    plt.box(False)
    plt.show()
```

## Plot test set data vs predictions

```python
# Prediction on test set
if not df.empty:
    plt.scatter(X_test, y_test, color = 'lightcoral')
    plt.plot(X_train, y_pred_train, color = 'firebrick') # The regression line is the same
    plt.title('Profit vs Sales (Test Set)')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.legend(['Pred(y_train)', 'y_test'], title = 'Prediction/Actual', loc='best', facecolor='white')
    plt.box(False)
    plt.show()
```

```python
# Regressor coefficients and intercept
if not df.empty:
    print(f'Coefficient: {regressor.coef_}')
    print(f'Intercept: {regressor.intercept_}')
```

# **Multiple Linear Regression - Global Superstore**

# Import data

# Analyze data

## Describe relevant features

```python
# We will use 'Sales', 'Discount', 'Quantity', and 'Profit'
if not df.empty:
    print(df[['Sales', 'Discount', 'Quantity', 'Profit']].describe())
```

## Distribution

```python
# Data distribution for the target variable 'Profit'
# This is the same as in the simple regression part, but crucial for this analysis too.
if not df.empty:
    plt.title('Profit Distribution Plot')
    sns.distplot(df['Profit'])
    plt.show()
```

## Relationship between Numerical Variables (Pair Plot)
A pair plot helps us visualize the relationships between all pairs of our numerical variables at once. We can see scatter plots for each pair and the distribution of each variable on the diagonal.

```python
# Select the numerical columns we plan to use, plus the target variable
if not df.empty:
    numerical_features = ['Sales', 'Discount', 'Quantity', 'Profit']
    sns.pairplot(df[numerical_features])
    plt.show()
```

## Relationship between Categorical Variables and Profit (Box Plots)
To see how our categorical predictors (`Segment` and `Category`) relate to `Profit`, box plots are very effective. They show the distribution of profit for each category.

```python
if not df.empty:
    # Relationship between Segment and Profit
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Segment', y='Profit', data=df)
    plt.title('Profit Distribution by Customer Segment')
    plt.show()

    # Relationship between Category and Profit
    plt.figure(figsize=(12, 7))
    sns.boxplot(x='Category', y='Profit', data=df)
    plt.title('Profit Distribution by Product Category')
    plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
    plt.tight_layout()
    plt.show()
```

# Split into Independent/Dependent variables

```python
# Split dataset in dependent/independent variables
if not df.empty:
    # X includes all our selected predictors
    features = ['Sales', 'Discount', 'Quantity', 'Segment', 'Category']
    X = df[features].values
    
    # y is the target variable
    y = df['Profit'].values
```

# One-hot encoding for categorical variables


```python
# One-hot encoding of categorical data
# 'Segment' (index 3) and 'Category' (index 4)
if not df.empty:
    ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3, 4])], remainder='passthrough')
    X = np.array(ct.fit_transform(X))
    print("Categorical data encoded successfully.\n")
```

# Split into Train/Test sets & Train model

```python
# Split dataset into test/train
if not df.empty:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train multiple regression model
if not df.empty:
    multi_regressor = LinearRegression()
    multi_regressor.fit(X_train, y_train)
    print("Model trained successfully.")
```

# Predict results

```python
# Predict result
if not df.empty:
    y_pred = multi_regressor.predict(X_test)
```

# Compare predictions

```python
# Compare predicted result with actual value
if not df.empty:
    np.set_printoptions(precision=2)
    result = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)
    result_df = pd.DataFrame(result, columns=['Predicted Profit', 'Actual Profit'])
    
    # Print the first 20 for a sample comparison
    print(result_df.head(20))
```
