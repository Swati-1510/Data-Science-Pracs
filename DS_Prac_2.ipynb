# Load Dataset

```python
import pandas as pd
import numpy as np

# --- 1. SETUP: LOAD YOUR DATA ---
FILE_NAME = 'kc_house_data.csv'

df = pd.read_csv(FILE_NAME)

```

# 2.1 Creating a Data Frame 

```python
print("\n--- Creating and Inspecting a DataFrame ---")
print("First 2 Rows (df.head(2)):")
print(df.head(2))
print("\nDimensions (df.shape): {}".format(df.shape))
print("\nDescriptive Statistics (df.describe()):")
print(df.describe())
```

# 2.2 Navigating DataFrame

```python
print("\n--- 2.2 Navigating DataFrame (iloc) ---")
# Select the first row (index 0)
print("\nFirst Row (df.iloc[0]):")
print(df.iloc[0])
# Select rows 1 up to (but not including) 4
print("\nRows 1 to 3 (df.iloc[1:4]):")
print(df.iloc[1:4])
```

# 2.4 Selecting Rows Based on Conditionals

```python
print("\n--- 2.4 Selecting Rows Based on Conditionals ---")
# Find houses with more than 4 bedrooms AND a grade higher than 10
CONDITION_1 = (df['bedrooms'] > 4)
CONDITION_2 = (df['grade'] > 10)

filtered_df = df[CONDITION_1 & CONDITION_2]

print("Houses with >4 Bedrooms AND Grade > 10 (First 2 Rows):")
print(filtered_df.head(2))
```

# 2.5 Replacing Values

```python
print("\n--- 2.5 Replacing Values ---")
# Replace 'view' value 0 with 'NoView' and 1 with 'HasView' (Single replacement)
view_series = df['view'].replace(0, 'NoView').head(5)
print("Single Replacement (View 0 -> NoView):")
print(view_series)

# Replace multiple values at once (Multiple replacement)
condition_series = df['condition'].replace([3, 5], ['Average', 'Excellent']).head(5)
print("\nMultiple Replacement (Condition 3->Average, 5->Excellent):")
print(condition_series)
```

# 2.6 Renaming Columns

```python
print("\n--- 2.6 Renaming Columns ---")
# Rename 'sqft_living' to 'Living_Area'
df_renamed = df.rename(columns={'sqft_living': 'Living_Area'}).head(2)
print("Renamed 'sqft_living' to 'Living_Area':")
print(df_renamed)
```

# 2.7 Finding Min, Max, Sum, Average (Mean), Count 

```python
print("\n--- 2.7 Finding Descriptive Statistics ---")
FEATURE = 'price'
print(f"Statistics for '{FEATURE}':")
print("Maximum: {:.2f}".format(df[FEATURE].max()))
print("Minimum: {:.2f}".format(df[FEATURE].min()))
print("Mean: {:.2f}".format(df[FEATURE].mean()))
print("Standard Deviation: {:.2f}".format(df[FEATURE].std()))
print("Count: {}".format(df[FEATURE].count()))
```

# 2.8 Finding Unique Values

```python
print("\n--- 2.8 Finding Unique Values ---")
FEATURE = 'bedrooms'
print(f"Unique values in '{FEATURE}':")
print(df[FEATURE].unique())
print(f"\nValue Counts for '{FEATURE}':")
print(df[FEATURE].value_counts().head(5))
```

# 2.9 Handling Missing Values (isnull) & 3.10 Deleting a Column 

```python
print("\n--- 2.9 Handling Missing Values (isnull) ---")
# The King County data is usually very clean, so we check for missing 'price'
print("Check for Missing 'price' values (First 5 Rows):")
print(df['price'].isnull().head(5))

print("\n--- 3.10 Deleting a Column (drop) ---")
# Drop the 'view' column (axis=1 for column)
df_dropped = df.drop('view', axis=1)
print("DataFrame after dropping 'view' column (First 2 Rows):")
print(df_dropped.head(2))
```

# 2.11 Deleting a Row

```python
print("\n--- 2.11 Deleting a Row (by index) ---")
# Delete the row at index 0 (the first row)
df_row_deleted = df.drop(0, axis=0)
print("DataFrame after dropping the first row (head shows the new first row):")
print(df_row_deleted.head(2))
```

# 2.16 Applying a Function Over All Elements in a Column 

```python
print("\n--- 2.16 Applying a Function (apply) ---")
# Define a simple function to categorize a house based on its living area
def categorize_area(sqft):
    if sqft < 1500: return 'Small'
    if sqft < 3000: return 'Medium'
    return 'Large'

area_category_series = df['sqft_living'].apply(categorize_area).head(5)
print("New Column from applying function to 'sqft_living' (First 5 Rows):")
print(area_category_series)
```

# 2.17 Applying a Function to Groups (groupby)

```python
print("\n--- 2.17 Applying a Function to Groups (groupby) ---")
# Group by 'bedrooms' and calculate the mean 'price' for each group
mean_price_by_bedroom = df.groupby('bedrooms')['price'].mean().head(5)
print("Mean Price grouped by Number of Bedrooms:")
print(mean_price_by_bedroom)
```

# 2.18 Concatenating DataFrames

```python
print("\n--- 2.18 Concatenating DataFrames (concat) ---")
# 1. Select all numerical columns
numerical_df = df.select_dtypes(include=[np.number])
# 2. Create a dummy categorical DataFrame (using the categorized area from 3.16)
categorical_df = pd.DataFrame(area_category_series)
categorical_df.columns = ['Area_Category']

# 3. Concatenate (join) them horizontally (axis=1)
# Note: Concatenation requires the two DataFrames to have the same indexes.
combined_df = pd.concat([numerical_df, categorical_df], axis=1)

print("Concatenated DataFrame (First 5 Rows - showing new 'Area_Category' column):")
print(combined_df.head(5))
```
